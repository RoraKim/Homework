## 복잡한 리스트의 합

> 2차원 리스트를 반복하는 방법을 알아봅시다. 


* 주어진 아래의 리스트를 반복하여 숫자의 합을 반환하시오.

---
**예시 입력**

```python
numbers = [
    [1, 4],
    [10, 5],
    [20, 30]    
]

```

**예시 출력**
```python
sum_list([[1, 4], [10, 5], [20, 30]]) # 70
```

### for 문을 활용하여 풀이하기

```python
for element in list:
```

````python
# 아래에 코드를 작성하세요.
def sum_list(numbers):
    add = 0 
    for i in numbers:
        for j in i:
            add += j
            
    return add
````

````python
print(sum_list([[1, 4], [10, 5], [20, 30]]))
````

### Index로 접근하여 풀이하기

```python
for element in range(len(list)):
```

### while 문을 활용하여 풀이하기

```python
while ____:
```

````python
## 아래에 코드를 작성하세요.
def sum_list_while(numbers):
    i = 0 
    result = 0 
    while i < len(numbers):
        j = 0
        while j < len(numbers[i]):
            #내부 요소의 값 result에 더하기 
            result += numbers[i][j]
            #내부 배열 값 하나 더하고 난 다음 
            #다음번째 값 더하려면 index 증가 
            j += 1
    
````

````python
print(sum_list_while([[1, 4], [10, 5], [20, 30]]))
````
